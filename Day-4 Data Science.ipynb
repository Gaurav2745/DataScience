## import class

## CLASSIFICATION

### decision tree classificatin

### import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
df
```

```python
#data set is given by social media website 
#some of the user click on ad and purchased the product
#some just ignored it
# 0 means purchased
# 1 means not purchased (seen the ad but not purchased)


# our task is to predict the customer will purchased or not purchased it
# age and salary are input variable 
# purchased is output variable
```

#### separate input and output variables

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

```python
x
```

```python
y.value_counts()
```

#### visualize the data

```python
plt.title('Purchase Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'],x['EstimatedSalary'] , c=y)
```

#### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

#### create the object

```python
classifier = DecisionTreeClassifier(random_state=0)
```

#### train the model

```python
classifier.fit(x,y)
```

#### Plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(25,25))
plot_tree(classifier,fontsize=7,feature_names=['age','sal'],class_names=['y','n'],filled=True , rounded=True);
```

#### predict on new data

```python
new = pd.DataFrame({
    'Age':[23,45,54,67],
    'EstimatedSalary':[45000,89000,23000,134000]
})
```

```python
new
```

```python
classifier.predict(new)
```

```python
#age of 23 , 45 will purchased the product 
```
