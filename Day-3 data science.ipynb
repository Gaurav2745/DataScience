## line plot

```python
x=np.array([3,11,16,23,27,34,48,55,56,67])
y=np.array([13,23,24,49,47,61,69,78,84,89])
```

```python
plt.plot(x,y)
#it used to plot the data in the line format
#it shows the relationship between x and y
#there are two parameter to the function 
#1->x-axis value
#2->y-axis value
```

```python
plt.title("plot-1")
plt.xlabel("value of x")
plt.ylabel('value of y')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.plot(x,y)
```

```python
plt.title("plot-2")
plt.xlabel("value of x")
plt.ylabel('value of y')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.xticks(range(0,81,10))#represented the value on x axis
plt.yticks(range(0,101,10))
plt.plot(x,y)
```

```python
z=np.array([87,77,75,71,54,33,21,23,19,10])
```

```python
plt.title("plot-3")
plt.xlabel("value of x")
plt.ylabel('value of y')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.xticks(range(0,81,10))#represented the value on x axis
plt.yticks(range(0,101,10))
plt.plot(x,z)
```

```python
plt.title("plot-4")
plt.xlabel("value of x")
plt.ylabel('value of y')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.xticks(range(0,81,10))#represented the value on x axis
plt.yticks(range(0,101,10))
plt.plot(x,z , marker='D' , color='red', ls=':' , lw=3 , mfc='green')
#marker : o > < v ^ D
```

```python
plt.title("plot-5")
plt.xlabel("value of x")
plt.ylabel('value of y')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.plot(x,y,label="2022",color='blue')
plt.plot(x,z,label="2023",color='red')
plt.legend()
#marker : o > < v ^ D
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
```

```python
plt.subplot(1,2,1) #one row two column and 1st column
plt.plot(x,y)
plt.subplot(1,2,2) #one row two column and 2nd row
plt.plot(x,z)
```

```python
plt.figure(figsize=(16,9))# use to set the aspect ratio of the plot
plt.title("plot-5")
plt.xlabel("value of x")
plt.ylabel('value of y')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.plot(x,y,label="2022",color='blue')
plt.plot(x,z,label="2023",color='red')
plt.legend()
```
